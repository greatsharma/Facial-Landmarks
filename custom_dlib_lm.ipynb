```
import dlib
import multiprocessing
```

```
options = dlib.shape_predictor_training_options()
```

```
# default options
options
```

```
train_file = "ibug_300W_large_face_landmark_dataset/labels_ibug_300W_train.xml"
test_file = "ibug_300W_large_face_landmark_dataset/labels_ibug_300W_test.xml"
```

```
%%time

options.be_verbose = True
options.cascade_depth = 15
options.tree_depth = 4
options.num_test_splits = 30
options.oversampling_amount = 5
options.oversampling_translation_jitter = 0.1
options.num_threads = multiprocessing.cpu_count()

save_file = "cd15_td4_split30_os5.dat"
dlib.train_shape_predictor(train_file, save_file, options)
```

```
dlib.test_shape_predictor(train_file, "cd15_td4_split30_os5.dat"), dlib.test_shape_predictor(test_file, "cd15_td4_split30_os5.dat")
```

```
%%time

options.be_verbose = True
options.cascade_depth = 20
options.tree_depth = 5
options.num_test_splits = 50
options.oversampling_amount = 10
options.oversampling_translation_jitter = 0.1
options.num_threads = multiprocessing.cpu_count()

save_file = "cd20_td5_split50_os10.dat"
dlib.train_shape_predictor(train_file, save_file, options)
```

```
%%time

options.be_verbose = True
options.cascade_depth = 25
options.tree_depth = 5
options.num_test_splits = 50
options.oversampling_amount = 20
options.oversampling_translation_jitter = 0.1
options.num_threads = multiprocessing.cpu_count()

save_file = "cd25_td5_split50_os20.dat"
dlib.train_shape_predictor(train_file, save_file, options)
```

```
dlib.test_shape_predictor(train_file, "cd25_td5_split50_os20.dat"), dlib.test_shape_predictor(test_file, "cd25_td5_split50_os20.dat")
```

```
# evaluate on default model
default_model = "drive/My Drive/shape_predictor_68_face_landmarks.dat"
dlib.test_shape_predictor(train_file, default_model), dlib.test_shape_predictor(test_file, default_model)
```
